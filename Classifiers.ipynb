```python
def Data_Split(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

    model1 = MultinomialNB()
    model1.fit(X_train,y_train)
    print("Naive Bayes Accuracy: ",model1.score(X_test, y_test))

    
    model2=LogisticRegression()
    model2.fit(X_train,y_train)
    print("Logistic Regression Accuracy: ",model2.score(X_test, y_test))

    model3 = KNeighborsClassifier(n_neighbors=6)
    model3.fit(X_train, y_train)
    print("Kneighbour Classifier's Accuracy :" + str(model3.score(X_test, y_test)))

    
    model4 = DecisionTreeClassifier()
    model4.fit(X_train, y_train)
    print("Decision-tree Classifier's Accuracy : " + str(model4.score(X_test, y_test)))
```

```python
def read_team_data(file_name):
    
    df = pd.read_csv(file_name)
    Y = df['match_winner']
    df.drop('match_winner', axis=1, inplace=True)

    X = pd.DataFrame()
    df['team1'] = df['team1'].astype('category')
    df['team2'] = df['team2'].astype('category')
    df['Ground'] = df['Ground'].astype('category')
    category_column = df.select_dtypes(['category']).columns
    X = pd.concat( [ X, df[category_column].apply(lambda ele: ele.cat.codes) ], axis = 1)
    
    X = pd.concat([ X, df[ 'first_batting' ] ], axis = 1)
    X = pd.concat([ X, df[ 'team1_bat_score' ] ], axis = 1)
    X = pd.concat([ X, df[ 'team1_bowl_score' ] ], axis = 1)
    X = pd.concat([ X, df[ 'team2_bat_score' ] ], axis = 1)
    X = pd.concat([ X, df[ 'team2_bowl_score' ] ], axis = 1)

    
  

    return X, Y
```

```python
X,y = read_team_data("final_ranking_score_1_.csv")
np.random.seed(0)
Data_split(X,y)
```
