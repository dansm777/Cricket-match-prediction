```python
import urllib.request, re
from html.parser import HTMLParser
import pandas as pd
```

```python
data = ""
```

```python
class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        global data
        if tag == "a":
            for name, value in attrs:
                if name == "href":
                    matched_data = re.search('/ci/engine/match/[0-9]+\.html', value)
                    if matched_data is not None:
                        #print(data)
                        data += str( matched_data.group(0) )  + "\n"
```

```python
url = "http://stats.espncricinfo.com/ci/engine/records/team/match_results.html?class=2;id="
```

```python
for yr in range(1995,2020):
    
    last_part = str(yr) + ";type=year"
    fP = open("links/links_"+ str(yr) + ".csv", "w")
    data = ""
    print(url + last_part)
    request = urllib.request.Request(url + last_part)
    response = urllib.request.urlopen(request).read()
    parser = MyHTMLParser()
    parser.feed(str(response))
    
    fP.write(data)
     # set for fast O(1) amortized lookup
    fP.close()

    df = pd.read_csv("links/links_"+ str(yr) + ".csv")
    print(len(df))
    df.drop_duplicates(subset=None, inplace=True)
    print(len(df))
    df.columns=['link']
    print(df)
    df.to_csv("links/links_"+ str(yr) + ".csv")
```
