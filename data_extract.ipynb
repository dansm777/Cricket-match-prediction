```python
files = list()
files = [f for f in li if isfile(join(mypath, f))]
len(files)
```

```python
yr=2014
for f in files:
    print(f)
    df = pd.read_csv("links/2014 to 2019/"+f)
    data = [] 
    un_open=[]
    #df=pd.read_csv(file_name)
    for i in range(len(df)):
        complete_url = url + df.iloc[i]['link']
        print(complete_url)
        try:
            print(i)
            df_li=[]
            df_li.append(complete_url)
            #print(df.iloc[i]['link'])
            #line.lstrip()

            request = urllib.request.Request(complete_url)
            page = urllib.request.urlopen(request).read()
            soup = BeautifulSoup(page, "html.parser")
            #print(soup.prettify())
            team=(soup.find_all('span', {'class': 'cscore_name cscore_name--long'}))
            df_li.append(str(team[0].string))
            df_li.append(str(team[1].string))
            print(team[0].string)
            print(team[1].string)
            stadium=(soup.find('div', {'class': 'stadium-details'}))
            df_li.append(str(stadium.text))
            #print(stadium.text)
            series_name=(soup.find('div', {'class': 'cscore_info-overview'}).text)
            li=series_name.split(",")
            df_li.append(str(li[-1]))
            df_li.append(str(li[-2]))
            #print(li[1])
            #print(li[2])
            toss=(soup.find_all('div', {'class': 'match-detail--item'}))
            if(toss[1].text.find(df_li[0])!=-1):
                df_li.append(1)
                if(toss[1].text.find('bat')!=-1):
                    df_li.append(1)
                else:
                    df_li.append(2)
            else:
                df_li.append(2)
                if(toss[1].text.find('bat')!=-1):
                    df_li.append(2)
                else:
                    df_li.append(1)


            print(toss[1].text)
            won=(soup.find('span', {'class': 'cscore_notes_game'}))
            if(won.text.find(df_li[0])!=-1):
                df_li.append(1)
            else:
                df_li.append(2)
            df_li.append(won.text)
            print((won.text))
            players = soup.find_all('div', {'class': 'flex-row'})
            cnt=0
            for i in range(len(players)):
                li=players[i].find_all('a')
                if (len(li)>=1):
                    if(',' in li[0].text.rstrip()):
                        for j in range(len(li)):
                            t=str(li[j].text.rstrip())
                            t=t.replace(',',"")
                            t=t.replace('(c)',"")
                            t=t.replace(' †',"")
                            cnt+=1
                            if(cnt<=22):
                                df_li.append(t)
                            #print(t)
                    else:
                        t1=str(li[0].text.rstrip())
                        t1=t1.replace('(c)',"")
                        t1=t1.rstrip().replace(' †',"")
                        cnt+=1
                        if(cnt<=22):
                            df_li.append(t1)
                        #print(t1)
            print(df_li)
            data.append(df_li) 
        except:
            un_open.append(complete_url)
            print("sleeping")
            for j in range(50,1,-1):
                time.sleep(1)
                print("sleep for ",j," seconds")
                
    op = pd.DataFrame(data)
    op.columns=['link','team1','team2','Ground','date','Series','toss_winner','first_batting','match_winner','Result','team1_player1','team1_player2','team1_player3','team1_player4','team1_player5','team1_player6','team1_player7','team1_player8','team1_player9','team1_player10','team1_player11','team2_player1','team2_player2','team2_player3','team2_player4','team2_player5','team2_player6','team2_player7','team2_player8','team2_player9','team2_player10','team2_player11']
    op.to_csv("op"+f)
    yr+=1
    print("sleeping")
    for j in range(50,1,-1):
        time.sleep(1)
        print("sleep for ",j," seconds")
```
